```python
import statsmodels.api as sm

nbt=len(bt)
x_multiple = [np.ones([nbt,1]),bt['CRIM'],bt['ZN'],bt['INDUS'],bt['CHAS'],bt['NOX'],bt['RM'],bt['AGE'],bt['DIS'],bt['RAD'],bt['TAX'],bt['PTRATIO'],bt['B'],bt['LSTAT']]
y = bt['MEDV'].values
X_multiple = np.column_stack(x_multiple)
ols =sm.OLS(y, X_multiple)
results = ols.fit()
print(results.summary())
```
